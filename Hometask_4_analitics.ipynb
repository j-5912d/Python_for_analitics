# Задание 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only = True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
sns.heatmap(corr_matrix, cmap = 'coolwarm')
```

По тепловой карте смотрим, что имеет кореляцию с ценой, приямую или обратную, это такие велечины как:
1. оценка (grade),
2. площади всех помещений жилая площадь корелируют со стоимостью (sqft_living) .
3. Вид из окон обладает средней кореляцией (view),
4. так же количество ванных комнат корелирует со стоимостью, но не сильно (bathrooms)


```python
sns.jointplot(x=df['sqft_living'], y=df['sqft_above'], kind='reg');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg');
```

Не совсем понятен смысл данного параметра (скорее всего какая-то площадь дома, высчитывающаяся относительно жилой площади) эта величина почти идеально корелирует с жилой площадью, поэтому не удивительно, что она корелирует и со стоимостью дома.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

как видно, всё в полне закономерно, чем больше жилая площадь помещения, тем дороже стоит дом. Соответственно это основной параметр влияющий на стоимость дома

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Как видно из графика распределение цены и оценки имеет закономерность, что подтверждает наличие пусть и не полностью точной но давольно таки неплохой кореляции между оценкой и стоимостью

```python
sns.jointplot(x=df['price'], y=df['yr_built'], kind='reg');
```

взаимные распределения между годом постройки и ценой никак не корелируют друг с другом и нет никакой взаимной связи

```python
sns.jointplot(x=df['price'], y=df['condition'], kind='reg');
```

```python
sns.jointplot(x=df['price'], y=df['condition'], kind='kde');
```

Не смотря на то, что состояние домов должна по логике вещей влиять на стоимость, в реальности дом в ужасных условиях может стоить больше, чем дом в отличных условиях - не корелирует

```python
df['long'].describe(), df['lat'].describe()
```

```python
sns.jointplot(x=df['long'], y=df['lat'], kind='reg');
```

Просто представление домов на карте согласно долготе и широте
